```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# -------------------------------
#from sklearn.datasets.load
from sklearn.datasets import load_breast_cancer
d = load_breast_cancer()
#type(d)
print(d.keys())
```

```python
# prepare input
x= pd.DataFrame(d['data'],columns=d['feature_names'])
x[:2]
```

```python
# output
y = d['target']
y[:2]
```

```python
x.shape
```

```python
# split the data into Training and Testing
from sklearn.model_selection import train_test_split
```

```python
a1,a2,a3,a4 = [200,300,400,500]
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,train_size=.8)  # 80-20
```

X_train,y_train--> training--> fit()

X_test,y_test--> testing --> predict(X_test)--> y_pred

```python
len(X_train)
```

1. Data Prepare
2. Data Analysis
3. Training
4. Testing
5. Validation

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model
```

```python
# Training
model.fit(X_train,y_train)
```

```python
X_test[:3]
```

```python
# Testing
y_pred = model.predict(X_test)  # predicted
y_pred
```

```python
y_test # actual result
```

```python
X_test[:3]
```

```python
model.score(X_train,y_train)*100  # data ka score
```

# **Check how good your model is**

```python
# confusion matrix- it has 4 quadrants
# TP True Positive
# TN True Negative
# FP False Positve
# FN False Negative
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# confusion_matrix(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(cm,annot=True,xticklabels=['malignant','benign'],yticklabels=['malignant','benign'])
```

```python
sum((y_test == 1) & (y_pred==0))
```

```python
# score range is 0-1
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
# accuracy tells u how accurate your model is
accuracy_score(y_test,y_pred)*100
```

```python
# precision- TPR True Positive Rate  TP/TP+FP
precision_score(y_test,y_pred)*100
```

```python
# recall- Sensivity of a data  TP/TP+FN
recall_score(y_test,y_pred)*100
```

```python
# f1_score- Harmonic Mean of both Precison and recall
f1_score(y_test,y_pred)*100
```

```python
## to get the combined result of all the metrices
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
